```python
%%time
import random
from math import exp, sqrt

S_0 = 100
r = 0.05
T = 1.0
sigma = 0.2

values = []

for _ in range(1000000):
    S_T = S_0 * exp((r - 0.5 * sigma ** 2) * T +
                    sigma * random.gauss(0, 1) * sqrt(T))
    values.append(S_T)
```

```python
%%time
import numpy as np

S_0 = 100
r = 0.05
T = 1.0
sigma = 0.2

value = S_0 * np.exp((r - 0.5 * sigma ** 2) * T +
                    sigma * np.random.standard_normal(1000000) * np.sqrt(T)) 
```
